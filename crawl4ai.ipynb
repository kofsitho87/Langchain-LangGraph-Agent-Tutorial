```python
import asyncio
from crawl4ai.deep_crawling import BFSDeepCrawlStrategy
from crawl4ai.content_scraping_strategy import LXMLWebScrapingStrategy

from crawl4ai import (
    AsyncWebCrawler, CrawlerRunConfig,
    BrowserConfig,
    CacheMode,
    CrawlerRunConfig,
    DefaultMarkdownGenerator,
)

# 브라우저 설정 - 안정성 향상을 위한 옵션 추가
browser_config = BrowserConfig(
    browser_type="chromium",
    headless=True,
)

# 크롤링 설정
crawler_run_config = CrawlerRunConfig(
    cache_mode=CacheMode.BYPASS,
    excluded_tags=["script", "style", "header", "footer", "input", "textarea"],
    # exclude_social_media_domains=["Facebook", "Twitter"],
    exclude_external_images=True,
    exclude_external_links=True,
    exclude_social_media_links=True,
    image_score_threshold=1,
    # image_description_min_word_threshold=0,
    excluded_selector="#header, #footer, #hd, #quick, #ft, #terms, .datepicker",
    markdown_generator=DefaultMarkdownGenerator(
        options={
            "ignore_links": True,
            "ignore_images": True,
        },
    ),
    deep_crawl_strategy=BFSDeepCrawlStrategy(
        max_depth=1, 
        max_pages=1,
        include_external=False,
    ),
    scraping_strategy=LXMLWebScrapingStrategy(),
    verbose=False
)

async def run_crawler():

    async with AsyncWebCrawler(config=browser_config) as crawler:
        results = await crawler.arun("http://www.hukins.co.kr/mn01_02.php", config=crawler_run_config)

        print(f"Crawled {len(results)} pages in total")

        # Access individual results
        for result in results:  # Show first 3 results
            print(f"URL: {result.url}")
            print(f"Depth: {result.metadata.get('depth', 0)}")

            print(result.media)
            print("--------------------------------")

await run_crawler()
```
